```python
#tradando dados da planilh de conexoes entre atores
origem=[]
destino=[]
for i in range(newdata.shape[0]):
    origem.append(int(str(newdata.iloc[i]).split("    ")[1].split(" ")[0]))
    tempo=str(newdata.iloc[i]).split("    ")[1].split(" ")[1]+" cabou"
    destino.append(int(tempo.split("Name:")[0]))
#conversao elenco em id

elencoid=[]
for i in elencoFilmes:
    listtemp = []
    for j in i:
        listtemp.append(int(mapatoresid[j]))
    elencoid.append(listtemp)
print("!")
```

```python
#preparando coluna de pesos para gerar o novo arquivo csv
peso=[]
countagem=0
relacoesxxx= []
stringtempxxx=""
for i in range(len(origem)):
    
    for j in elencoid:
        
        if (origem[i] in j) and (destino[i] in j):
            countagem+=1
            
    peso.append(countagem)    
    countagem=0    

print("1")              
  
df3={'SOURCE': origem, 'TARGET': destino, 'PESO':peso}
dfinal2 = pd.DataFrame(data=df3)
#gerando o arquivo final 
dfinal2.to_csv('tabela2.csv', encoding='utf-8', index=False)
```
